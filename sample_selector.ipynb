```python
import pandas as pd

real_df = pd.read_csv("alz-speech.csv")
real_df["word_count"] = real_df["transcript"].str.split().str.len()

def sample_varied(df, n=5, seed=42):
    df = df.copy()
    df["age_bin"]    = pd.qcut(df["word_count"], q=n, duplicates="drop", labels=False)
    # Sort by age then word_count alternating to spread diversity, then sample
    df = df.sort_values(["age_bin", "sex", "age"]).reset_index(drop=True)
    step = max(1, len(df) // n)
    indices = list(range(0, len(df), step))[:n]
    return df.iloc[indices][["subject", "sex", "age", "word_count", "transcript"]].reset_index(drop=True)

ad_examples   = sample_varied(real_df[real_df["ad"] == 1], n=5, seed=42)
ctrl_examples = sample_varied(real_df[real_df["ad"] == 0], n=5, seed=99)

print("=" * 70)
print("  AD EXAMPLES (label=1)")
print("=" * 70)
for _, row in ad_examples.iterrows():
    print(f"\n[subject={row['subject']} | sex={row['sex']} | age={row['age']} | words={row['word_count']}]")
    print(row["transcript"])

print("\n" + "=" * 70)
print("  CONTROL EXAMPLES (label=0)")
print("=" * 70)
for _, row in ctrl_examples.iterrows():
    print(f"\n[subject={row['subject']} | sex={row['sex']} | age={row['age']} | words={row['word_count']}]")
    print(row["transcript"])
```
